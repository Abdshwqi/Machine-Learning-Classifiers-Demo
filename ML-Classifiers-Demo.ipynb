```python
class KNN:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        y_pred = []
        for x in X:
            #   Calculate the distance between points wit every point in a training
            distances = np.sqrt(np.sum((self.X_train - x)**2, axis=1))
            # sort distances and return indices of the sorted array
            indices = np.argsort(distances)
            #  get the labels Knn
            k_neighbors = [self.y_train[i] for i in indices[:self.k]]
            # Return the most popular vote
            y_pred.append(max(k_neighbors, key = k_neighbors.count))
        return np.array(y_pred)


knn_own = KNN(3)
knn_own.fit(X_train, y_train)
y_pred_own = knn_own.predict(X_test)
print("Predictions usnig mine KNN class: ",y_pred_own)
```

```python
accuracy_own = accuracy_score(y_test, y_pred_own)
print("Accuracy usnig mine KNN class with K=3: {:.2f}%".format(accuracy_own * 100))
```

```python
import numpy as np

class LogisticRegressor:
    def __init__(self, learning_rate=0.001, max_iter=1000):
        self.learning_rate = learning_rate
        self.max_iter = max_iter

    def sigmoid(self, x):
        # map any real value between 0 and 1
        return 1 / (1 + np.exp(-x))

    def fit(self, X, y):
        # Initialize the parameters
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        # Perform the gradient descent
        for i in range(self.max_iter):
            z = np.dot(X, self.weights) + self.bias
            a = self.sigmoid(z)
            e = y - a
            dw = np.dot(X.T, e)
            db = np.sum(e)
            self.weights += self.learning_rate * dw
            self.bias += self.learning_rate * db

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        a = self.sigmoid(z)
        return np.round(a)

# Instantiate the logistic regression model
log_reg = LogisticRegressor(learning_rate=0.01, max_iter=1000)

# Fit the model to the training data
log_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = log_reg.predict(X_test)


# prediction results
print("Prediction results: ", y_pred)
print("OWN Accuracy : {:.2f}%".format(accuracy_own * 100))
```
